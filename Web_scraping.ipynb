````python
import bs4
from bs4 import BeautifulSoup
import csv
import requests
import time
import pandas as pd
import urllib
import re
import pickle
````

````python
yr_mon_list=[]
def getnums(s, e,i):
   return list(range(s, e,i))

yrs = getnums(2009,2019,1)

for n in yrs:
    for month in range(1,13):
        yr_mon_list.append(int(str(n)+str(month).rjust(2, '0')))

yr_mon_list = yr_mon_list[:-2]     # Remove nov,dec of 2018    
print(yr_mon_list)
````

````python
#### Start you code here, you are free to add any number of cells

baseURL = "http://www.estesparkweather.net/archive_reports.php?date="

column_names = ['Average temperature (°F)', 'Average humidity (%)',
 'Average dewpoint (°F)', 'Average barometer (in)',
 'Average windspeed (mph)', 'Average gustspeed (mph)',
 'Average direction (°deg)', 'Rainfall for month (in)',
 'Rainfall for year (in)', 'Maximum rain per minute',
 'Maximum temperature (°F)', 'Minimum temperature (°F)',
 'Maximum humidity (%)', 'Minimum humidity (%)', 'Maximum pressure',
 'Minimum pressure', 'Maximum windspeed (mph)',
 'Maximum gust speed (mph)', 'Maximum heat index (°F)']

df=pd.DataFrame()

for yrmon in yr_mon_list: #range(201012,201013): 
    print(baseURL+str(yrmon))
    URL = baseURL+str(yrmon)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="main-copy")
    table_elements = results.find_all("table")

    for i in range(0,len(table_elements)):
        data = pd.read_html(str(table_elements))[i]
        day = data.iloc[0][0].split(' ')[1]
        if(day.isnumeric()):
            data_pivot = data[1:].T.iloc[1:,:]
            data_pivot.columns = column_names
            data_pivot.index = [str(yrmon)[0:4]+'-'+str(yrmon)[4:]+'-'+day.rjust(2, '0')] 
            df=df.append(data_pivot)
    
````

````python
df2 = df.iloc[:-3,:]
df2.tail(5)
````

````python
def clean_up(el):
    #print(type(el))
    x = str(el)
    #print(type(x))
    if 'in.' in x:
        return float(x.split('in.')[0])
    elif '%' in x:
        return int(x.split('%')[0])
    elif 'mph' in x:
        return float(x.split('mph')[0])
    elif '°F' in x:
        return float(x.split('°F')[0])
    elif '°' in x:
        return int(x.split('°')[0])
    
    return x
    
````

````python
data_clean = df2.applymap(clean_up)
data_clean
````

````python
data_clean.index
````

````python
data_clean.index = pd.to_datetime(data_clean.index)
````

````python
import pickle
with open("dataframe.pk", "wb") as file:
    pickle.dump(data_clean, file)
````
